```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'AnnualSpending': [1200, 3000, 450, 800, 2200],
    'PurchaseFrequency': [12, 30, 5, 8, 25],
    'AvgOrderValue': [100, 100, 90, 100, 88]
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['AnnualSpending', 'PurchaseFrequency', 'AvgOrderValue']]
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
```

```python
# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_
```

```python
# Print results
print("Cluster Centers:\n", kmeans.cluster_centers_)
print(f"Inertia: {kmeans.inertia_:.2f}")
print("\nClustered Data:\n", df)
```

```python
# 3D Scatter Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    df['AnnualSpending'],
    df['PurchaseFrequency'],
    df['AvgOrderValue'],
    c=df['Cluster'],
    cmap='viridis',
    s=100
)
```

```python
ax.set_xlabel('Annual Spending')
ax.set_ylabel('Purchase Frequency')
ax.set_zlabel('Avg Order Value')
ax.set_title('Customer Segmentation Using K-Means')
plt.colorbar(scatter)
plt.show()
```
