# **Project Summary -**

Mental Health in Tech Survey


The “Mental Health in Tech Survey” project is based on a comprehensive dataset collected in 2014, aiming to explore the status and perceptions of mental health in the technology sector. As mental health becomes an increasingly important topic in modern workplaces, especially in high-stress environments like the tech industry, this survey sheds light on the challenges employees face, workplace responses, and the overall culture surrounding mental well-being.

1. Background and Purpose
The dataset, originally sourced from the organization Open Sourcing Mental Illness (OSMI), includes responses from individuals working in technology-related fields. The core objective of the survey is to assess how mental health is perceived, discussed, and supported in tech workplaces. It also aims to analyze factors that contribute to or hinder employee well-being, such as employer-provided benefits, company culture, and openness around discussing mental health.



2. Dataset Overview
The dataset contains multiple variables, including demographic data (age, gender, country, U.S. state), employment details (self-employed status, company size, tech company or not), and mental health-related variables (family history, treatment-seeking behavior, work interference, etc.).

Key columns include:

* Mental health history: personal and family history of mental illness.

* Treatment: whether respondents have sought treatment.

* Work interference: how mental health conditions impact work performance.

* Support system: employer benefits, wellness programs, anonymity, and leave policies.

Perceptions: how seriously mental health is taken compared to physical health, consequences of disclosure, willingness to discuss issues with coworkers, supervisors, or potential employers.

3. Insights and Observations
Family History and Treatment Seeking: A significant number of respondents who have a family history of mental illness were more likely to seek treatment themselves, indicating a strong correlation between awareness and action.

Work Interference: Many individuals reported that mental health conditions interfere with their work, yet few workplaces offer effective systems to address these challenges.

Company Support: A considerable number of companies in the tech sector still lack structured mental health support systems, including benefits, wellness programs, or even basic awareness initiatives.

Anonymity and Leave: Concerns around privacy and the fear of professional consequences discourage employees from seeking help. Furthermore, medical leave for mental health is often perceived as less accessible or less acceptable compared to physical health leave.

Stigma in the Workplace: There remains a notable stigma associated with mental health. Many respondents fear negative outcomes from disclosing mental health issues to employers or during interviews.

Willingness to Discuss: While a fair number are open to discussing physical health with employers, fewer are comfortable discussing mental health. This highlights a gap in psychological safety and workplace openness.

4. Relevance and Implications
This study is especially important in today's context where remote work, burnout, and mental stress have become prevalent. The findings suggest that even in progressive industries like tech, mental health awareness, education, and support need significant improvement.

Employers should:

 * Offer mental health benefits.

 * Conduct wellness programs focusing on mental well-being.

 * Ensure confidentiality and non-retaliation for those seeking help.

 * Create a supportive work environment where discussions around mental health are normalized.

5. Conclusion
The “Mental Health in Tech” survey reveals deep insights into the mental health climate of the tech industry. It emphasizes the urgent need for policy improvements, cultural change, and awareness programs to better support employees. Organizations that prioritize mental health not only enhance employee satisfaction and productivity but also create more resilient and innovative workforces. This project underscores the power of data in driving meaningful workplace reform and fostering a more compassionate professional environment.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd

# Load the dataset
df = pd.read_csv("survey.csv")

# 1️ Remove duplicate rows
df.drop_duplicates(inplace=True)

# 2️ Filter invalid Age values (only keep Age between 16 and 100)
df = df[(df['Age'] >= 16) & (df['Age'] <= 100)]

# 3️ Normalize Gender column
df['Gender'] = df['Gender'].str.lower().str.strip()
df['Gender'] = df['Gender'].replace({
    'male': 'male', 'm': 'male', 'man': 'male', 'cis male': 'male',
    'female': 'female', 'f': 'female', 'woman': 'female', 'cis female': 'female',
    'trans female': 'other', 'trans male': 'other', 'non-binary': 'other',
    'genderqueer': 'other', 'agender': 'other', 'other': 'other',
    'queer': 'other', 'fluid': 'other', 'genderfluid': 'other',
    'male-ish': 'male', 'female (trans)': 'other', 'femake': 'female',
    'malr': 'male', 'cis man': 'male', 'cis woman': 'female'
})

# 4️ Fill missing/null values
df['self_employed'].fillna('No', inplace=True)
df['work_interfere'].fillna("Don't know", inplace=True)
df['state'].fillna('Unknown', inplace=True)

# 5️ Drop irrelevant or mostly empty columns
df.drop(['Timestamp', 'comments'], axis=1, inplace=True)

#  Final dataset ready for analysis
print(" Cleaned dataset shape:", df.shape)
print(df.head())

# Fix for fillna FutureWarning
df['self_employed'] = df['self_employed'].fillna('No')
df['work_interfere'] = df['work_interfere'].fillna("Don't know")
df['state'] = df['state'].fillna('Unknown')



````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Step 1: Select relevant categorical columns related to mental health support
cols_of_interest = [
    'treatment', 'family_history', 'remote_work', 'tech_company',
    'benefits', 'care_options', 'wellness_program', 'seek_help',
    'anonymity', 'leave', 'mental_health_consequence',
    'phys_health_consequence', 'mental_health_interview',
    'phys_health_interview', 'mental_vs_physical', 'obs_consequence'
]

# Step 2: Extract subset and handle missing values
df_corr = df[cols_of_interest].fillna("Unknown").copy()

# Step 3: Encode categorical values into numeric using LabelEncoder
le = LabelEncoder()
for col in df_corr.columns:
    df_corr[col] = le.fit_transform(df_corr[col].astype(str))

# Step 4: Compute correlation matrix and plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(df_corr.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Mental Health Support Variables", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
````

To prepare this chart, categorical variables were converted into numeric form using Label Encoding, and then the correlation matrix was calculated using .corr(). The seaborn heatmap visually represents these correlations using a diverging color palette to emphasize both positive and negative relationships.
